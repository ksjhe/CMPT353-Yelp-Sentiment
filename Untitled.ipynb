```python
df = pd.read_parquet('data/edmonton_cleaned.parquet', engine='auto')
df.head(5)
```

```python
food_categories = ['Bagels', 'Bakeries', 'Barbeque', 'Bars', 'Beer', 'Beer Bar', 'Bistros', 'Breakfast & Brunch', 'Breweries', 'Brewpubs', 'Bubble Tea', 'Buffets',
                   'Burgers', 'Cafes', 'Cheesesteaks', 'Chicken Wings', 'Chocolatiers & Shops', 'Coffee & Tea', 'Coffee Roasteries', 'Comfort Food', 'Creperies', 'Desserts', 'Dim Sum', 'Diners', 'Dinner Theater', 'Distilleries', 'Dive Bars', 'Donairs', 'Donuts','Dumplings', 'Ethnic Food', 'Falafel', 'Fast Food', 'Fish & Chips', 'Fondue', 'Food', 'Food Court', 'Food Delivery Services', 'Food Stands', 'Food Trucks', 'Gay Bars', 'Gelato', 'Hong Kong Style Cafe', 'Hot Dogs', 'Hot Pot', 'Ice Cream & Frozen Yogurt', 'Imported Food', 'Irish Pub', 'Izakaya', 'Juce Bars & Smoothies', 'Live/Raw Food','Macarons', 'Pancakes', 'Patisseries/Cake Shop','Pizza', 'Poke', 'Pubs', 'Ramen','Restaurants', 'Salad', 'Sandwiches', 'Seafood', 'Soul Food', 'Speakeasies', 'Sports Bar', 'Steakhouses', 'Sushi Bars', 'Tacos', 'Tapas Bar', 'Tapas/Small Plates', 'Tex-Mex', 'Themed Cafes', 'Waffles','Whiskey Bars', '']



df['categories'] = df['categories'].astype(str)

df['filter'] = df['categories'].apply(lambda x: ''.join([L for L in food_categories if L in x]))
df['categories'] = df['filter'].mask(df['filter'] == '')
```
