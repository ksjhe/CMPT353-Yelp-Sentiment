```python
# finds all unique categories, appends them, and sorts
all_categories = df['categories'].unique().tolist()
all_categories = ', '.join(all_categories)
all_categories = all_categories.split(', ')
all_categories = list(set(all_categories))
all_categories.sort()

all_categories
```

```python
food_categories = ['Bagels', 'Bakeries', 'Barbeque', 'Bars', 'Beer', 'Beer Bar', 'Bistros', 'Breakfast & Brunch', 'Breweries', 'Brewpubs', 'Bubble Tea', 'Buffets',
                   'Burgers', 'Cafes', 'Cheesesteaks', 'Chicken Wings', 'Chocolatiers & Shops', 'Coffee & Tea', 'Coffee Roasteries', 'Comfort Food', 'Creperies', 'Desserts', 'Dim Sum', 'Diners', 'Dinner Theater', 'Distilleries', 'Dive Bars', 'Donairs', 'Donuts','Dumplings', 'Ethnic Food', 'Falafel', 'Fast Food', 'Fish & Chips', 'Fondue', 'Food', 'Food Court', 'Food Delivery Services', 'Food Stands', 'Food Trucks', 'Gay Bars', 'Gelato', 'Hong Kong Style Cafe', 'Hot Dogs', 'Hot Pot', 'Ice Cream & Frozen Yogurt', 'Imported Food', 'Irish Pub', 'Izakaya', 'Juce Bars & Smoothies', 'Live/Raw Food','Macarons', 'Pancakes', 'Patisseries/Cake Shop','Pizza', 'Poke', 'Pubs', 'Ramen','Restaurants', 'Salad', 'Sandwiches', 'Seafood', 'Soul Food', 'Speakeasies', 'Sports Bar', 'Steakhouses', 'Sushi Bars', 'Tacos', 'Tapas Bar', 'Tapas/Small Plates', 'Tex-Mex', 'Themed Cafes', 'Waffles','Whiskey Bars', '']


#categories is in some kind of object? converted to string for extraction.
df['categories'] = df['categories'].astype(str)

#masker for only restaurant values
masker = []

#using lambda function for filtering only keywords in food_categories list.
masker = df['categories'].apply(lambda x: ''.join([L for L in food_categories if L in x]))
df['categories'] = df['categories'].mask(masker == '')
```

```python
#remove all NaN values.
df = df.dropna()
```

# Exploring cleaned data

```python
#checking reviews per restaurant
count = df.groupby(by=["name"]).count().reset_index()

#putting in a count DF
count = count[['name','review_count']]

#just checking
print(count['review_count'].sum())

#checking reviews range for a unique restaurant
print(count.min())
print(count.max())
```
